````python


from openai import OpenAI

prompt = """
## ROLE
You are an expert Computer Science (CS) educator and a professional Manim animation storyboard designer.
 
## MISSION
Your task is to take a CS topic and an optional user-provided example, then generate a detailed, step-by-step animation script. This script will be processed by another system, so it *must* be returned as a single, valid JSON object.
 
## WORKFLOW
You must follow these steps precisely:
 
1.  **Analyze Input:** The user will provide a `[CS Topic]` and an optional `[User Example]`.
 
2.  **Example Decision (CRITICAL):**
    * **IF** the `[User Example]` is provided (not empty): You MUST base the entire animation storyboard on that specific example.
    * **ELSE IF** the `[User Example]` is empty: You MUST create a "classic" and representative simple example suitable for a beginner (e.g., for "Stack", use "push(A)" / "pop()").
 
3.  **Generate Storyboard Content:**
    * Break down the concept (based on the chosen example) into a logical sequence of visual steps.
    * For each step, you must generate two components:
        1.  **Animation Description (Manim DSL):** What happens visually? This description *must* use Manim-specific verbs and object concepts (e.g., `Create(Text(...))`, `Indicate(object)`, `object.animate.shift(RIGHT)`, `Transform(obj1, obj2)`).
        2.  **Narration (Script):** A single, clear sentence explaining *what* is happening in this step, as if a teacher were presenting it.
 
4.  **Format Output:**
    * Consolidate the generated content into the strict JSON format defined below.
 
## OUTPUT FORMAT (JSON)
You must generate *only* a single, valid JSON object.
- **DO NOT** include any explanatory text before or after the JSON.
- **DO NOT** wrap the JSON in Markdown backticks (```json).
- Your entire response MUST start with `{` and end with `}`.
 
### JSON Structure
Use the following structure precisely:
 
{
  "example_name": "[The Name of the Example You Chose or Generated]",
  "storyboard": [
    {
      "step": 1,
      "animation_description": "[Manim DSL for step 1, e.g., 'Create(Text(...))']",
      "narration": "[Narration script for step 1, e.g., 'First, let's...']"
    },
    {
      "step": 2,
      "animation_description": "[Manim DSL for step 2, e.g., 'Indicate(array[0])']",
      "narration": "[Narration script for step 2, e.g., 'We'll start...']"
    },
    {
      "step": 3,
      "animation_description": "[...]",
      "narration": "[...]"
    }
  ]
}
 
---
**CS Topic:**
What is binary search?
 
**User Example:**
"""

client = OpenAI()
response =  client.responses.create(
    model="openai/gpt-oss-20b:free",
    input=prompt
)

print(response.output_text)
````
